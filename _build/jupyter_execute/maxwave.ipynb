# Maximum wave profile

We will first examine the situation near the peak of a wave profile, which we place at $z=0$. At the peak, we already know that $v=0$. Stokes speculated in {cite}```Stokes_2009a``` appendix B, and {cite}`Stokes_2009b` that the maximum height wave would have a discontinuous slope with an angle of 120&deg; or $\pi/3$ radians. That such a behavior is admitted in the idealized theory was proved rigorously only in the middle of the 20th century.

We are looking for singular behaviour near $z=0$ to give a discontinuous wave profile. Since $z=0$ is where we have positioned our peak, we already know that the vertical velocity $v=0$ there.

Assuming that the gravitational potential is measured relative to this point, and that atmospheric pressure is absorbed into the Bernoulli constant, we have $R=\frac12u^2\ge0$. Elsewhere we define the head function $R=\frac12(u^2+v^2)+y$, so contour lines of constant $R$ give isobars, since $R+p/\rho=\mathrm{const.}$ by Bernoulli's principle. We have taken units where $g=1$ for convenience. We could also take $\rho=1$, but we will not be much concerned with pressure, except for the surface, where it is normalized to 0.

It is therefore interesting to study the possibility $R=0$ on the wave surface. Then the horizontal fluid velocity $u$ at the peak is 0 ($w=0$), remembering that we are moving with the wave peak so that the underlying fluid (deep down) is flowing at $c$ in the direction opposite to the travelling wave direction.

We assume near the peak, the surface declines in a linear fashion: $\eta=-ax$ when $x>0$ (the slope is positive for $x<0$), so $R=\frac12|w|^2+\eta=0$. Thus $|w|=\sqrt{-2\eta}=\sqrt{2ax}$.

Near the peak we also need the $u,v$ to be such that the flow direction is along $\eta$, thus $v=-au$. So $|w|^2=(1+a^2)u^2=-2\eta=2ax$. So $u=\sqrt{2ax/(1+a^2)}$, and $v=-a\sqrt{2ax/(1+a^2)}$

Now we want $w=u-iv$ to be an analytic function of $z=x+iy$. First $w=(1+ia)\sqrt{2ax/(1+a^2)}=\sqrt{2a(1+ia)x/(1-ia)}$.

On the surface $z=x+i\eta=(1-ia)x$, so $w=\sqrt{2a(1+ia)z/(1-ia)^2}$ seems to be a suitable continuation beyond the surface.

We expect $w$ to be real on the imaginary axis ($v=0$), so $w=A\sqrt{iz}$. The factor $i$ gives a positive real on the negative part of the imaginary axis ($i(iy)=-y$), so the branch cut is on the positive part of the imaginary axis, which is outside our water.

The point $z=0$ is the branch point of $\sqrt{iz}$. Above it on the positive imaginary axis, on one side $\sqrt{-y}=+i\sqrt{y}$ and on the other $\sqrt{-y}=-i\sqrt{y}$. Crossing the branch cut takes one to the second Riemann sheet of $\sqrt{iz}$, giving the negative values of the square root function. I hope with a bit of search engineering the reader will understand the various dropped buzzwords.

The surface near $z=0$ should have the form $r\exp(-i\theta)$ with $0<\theta<\pi/2$ ($x>0$). At the same time $w=t\exp(+i\theta)=A(iz)^\frac12=A(r\exp(i(\pi/2-\theta))^{1/2}$.

So $t=A\sqrt{r}$, and $\theta=\pi/4-\theta/2$, or solving $\theta=\pi/6$.

Now $R=A^2r/2-r\sin\theta=0$, and since $\sin\pi/6=1/2$, $A^2=1$.

Similar considerations for $x<0$ give a rise to the crest also at an angle $\pi/6$, so the angle between the rise and fall is $2\pi/3$, or 120&deg;.

A more algebraic approach, comparing $w=A\sqrt{iz}=\sqrt{2a(1+ia)z/(1-ia)^2}$, leads to the same conclusion. Squaring, and cancelling $z$:

$$A^2i=\frac{2a(1+ia)}{(1-ia)^2}$$

thus:

$$A^2(1-a^2-2ia)i=2a+2ia^2$$

The real part gives $2A^2a=2a$ and therefore $A^2=1$. The imaginary part becomes $1-a^2=2a^2$, which gives $a=1/\sqrt3$. This is the gradient of the slope, and the angle below the horizontal $\theta=\arctan(1/\sqrt3)=\pi/6=30^\circ$. The angle from the vertical is thus $\pi/3$, giving a peak angle of $2\pi/3$.

We can give a contour plot of our $R$ function:

````python
import matplotlib.pyplot as plt
import numpy as np

# grid

N=200
X,Y=np.mgrid[-1:1:N*1j,-1:1:N*1j]

Z=X+1j*Y

W=np.sqrt(1j*Z)
R=abs(W**2)/2+Y

fig, ax  = plt.subplots()
cr = ax.contour(X,Y,R,levels=np.linspace(-2,2,11),colors='blue')
ax.clabel(cr, fontsize=10)
plt.show()
````

We note the $R=0$ contour has the specified kink of $2\pi/3$ angle.

The complex potential corresponding to this flow, based at 0, is $\chi=\int_0^zwdz=2Ai^{1/2}z^{3/2}/3=2wz/3$. The stream function $\psi=\Im\chi$. These contours are shown in blue.

````python
C=2*W*Z/3
PSI=C.imag

cp = ax.contour(X,Y,PSI,levels=np.linspace(-2,2,11),colors='orange')
ax.clabel(cp, fontsize=10)
fig
````

We note the 0 contours of $R$ and $\psi$ match. The others do not, of course.
