# Testing module deepmax

```python
# global imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root

from deepmax.cooker import X_N,cooker,set_n,P,R
from deepmax.empirical import Ye
from deepmax.plot import grid,contours
```

```python
# using mult_n basis
from deepmax.mult_n import chi_n,w_n,C_n
set_n(w_n,chi_n,C_n)

N=27
X=X_N(N,alpha=1.25)
Y0=Ye(X)
Yc=cooker(Y0,X)
Hc=-Yc[-1]
Hc/2/np.pi
```

```python
Z=X+1j*np.array([0,*Yc])
C=C_n(Z)
```

```python
Grid=grid(0,np.pi-1j)
Z=Grid[0]+1j*Grid[1]
```

```python
R0=R(0,C)
PatZ=P(Z,C)
RatZ=R(Z,C)
CP=contours(Grid,PatZ,np.linspace(-1,+1,11),'blue')
CR=contours(Grid,RatZ,np.linspace(R0-1,R0+1,11),'orange')
plt.title(f'P=0,R={R0:8.4f}')
plt.show()
```

```python
X=np.linspace(0,np.pi,200)
```

```python
def FR(Y):
        Z=X+1j*Y
        return R(Z,C)

def FP(Y):
        Z=X+1j*Y
        return P(Z,C)
```

```python
YR=root(FR,-np.ones(X.shape)).x
YP=root(FP,-np.ones(X.shape)).x
```

```python
plt.plot(X,YP)
plt.plot(X,YR)
```

```python
plt.plot(X,YP-YR)
```

```python
max(abs(YP-YR))
```
